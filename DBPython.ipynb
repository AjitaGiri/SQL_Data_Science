```python
import mariadb
```

```python
from dotenv import load_dotenv
import os
```

```python
load_dotenv()

#connection parameters

conn_params={
    'host':os.getenv('DB_HOST'),
    'user':os.getenv('DB_USER'),
    'password':os.getenv('DB_PASSWORD'),
    'database':os.getenv('DB_NAME'),
    'port':3306
}

#create the connection
conn=mariadb.connect(**conn_params)

#create connection object
cur=conn.cursor()
```

```python
result=cur.execute('Select * from customers')
```

## insert data into database using python

```python
with conn.cursor() as cur:
    cur.execute('Insert into customers(id,first_name,country,score) values (?,?,?,?)', (6,'Shiva','Nepal',505))
    conn.commit()
```

## inserting multiple values at a time

```python
customer_details=[
    (7,'Lucky','UK',599),
    (8,'Sunoo','Korea',600)
]
```

```python
with conn.cursor() as cur:
 cur.executemany('Insert into practice_db.customers(id,first_name,country,score) values (?,?,?,?)',customer_details)
 conn.commit()
```

## updating the data

```python
with conn.cursor() as cur:
    cur.execute('UPDATE customers set country=? where id=?',('USA',8))
    conn.commit()
```

## deleting the data

```python
with conn.cursor() as cur:
    cur.execute('Delete from customers where id = ?',(7,))
    conn.commit()
```

## reading data

```python
with conn.cursor() as cur:
    cur.execute('Select * from customers')
    rows=cur.fetchall()

for row in rows:
    print(row)
```
